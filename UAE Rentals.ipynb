📌 Introduction
</h2>

This project analyzes Dubai's residential rental market using a dataset containing approximately 10,000 rental property records from January 2020 to April 2026. The data covers multiple communities across Dubai and captures rental market trends during a period marked by significant events, including the COVID-19 pandemic, Expo 2020, increased foreign investment, the Golden Visa expansion, and recent regional geopolitical tensions involving missile and drone attacks in the Gulf region.

The objective of this analysis is to explore rental price patterns, identify key factors influencing rents, and gain insights into Dubai's evolving rental market through data visualization and exploratory data analysis.

Derived Columns
</h1>

##### Creating New Column Based on Time Period

```python
df["Event_Period"] = np.where(df["Date_Listed"] <= '2020-03-01', "Pre-COVID",
    np.where(df["Date_Listed"] <= '2021-10-01', "COVID-19 Pandemic",
        np.where(df["Date_Listed"] <= '2022-04-01', "Expo 2020 Dubai",
            np.where(df["Date_Listed"] <= '2023-01-01', "Post-Expo Recovery",     
                np.where(df["Date_Listed"] <= '2026-02-01', "Golden Visa & Foreign Investment Growth",
                                                             "Regional Geopolitical Tensions")
))))
```

An event_period column was created by segmenting date_listed into key UAE market phases such as Pre-COVID, COVID-19, Expo 2020, Post-Expo Recovery, and subsequent investment and geopolitical periods.

##### Creating Year and Month Column

```python
df['Year'] = df['Date_Listed'].dt.year
df['Month'] = df['Date_Listed'].dt.month
```

#### Creating a Metro Accessibility Zone Column

```python
df["Metro_Accessibility"] = np.where(df["Metro_Distance_Min"] <= 10,"High Accessibilitty",
                            np.where(df["Metro_Distance_Min"]<=20,"Medium Accessibility","Low Accessibility"))
```

Rental Price Overview

#### Rent Distribution

```python
df["Annual_Rent_Usd"].describe()
```

***Insight:***

 The rental prices are not evenly distributed and are skewed to the right. A few very expensive luxury properties increase the average rent. However, there are some very high-priced properties (up to 1.17M USD), showing that Dubai has a strong luxury real estate market. Overall, the dataset has a wide variation in rental prices across different property types.

#### Higher Average Rent by Zone

```python
df.groupby("Zone")["Annual_Rent_Usd"].mean().sort_values(ascending = False).head(5)
```

***Insight:***

These zones are likely influenced by:
Prime location (city center / waterfront / luxury communities)
Better infrastructure and amenities
Higher demand from expats and high-income tenants.

#### Lower Average Rent by Zone

```python
df.groupby("Zone")["Annual_Rent_Usd"].mean().sort_values(ascending = True).head(5)
```

***Insight:***

 The top 5 zones show significantly lowerr average rental prices, indicating affordable residential areas.
Higher demand from expats and low-income tenants.

#### Average Rent by Community

```python
df.groupby("Community")["Annual_Rent_Usd"].mean().sort_values(ascending = False).head(5)
```

***Insight:***

 The top 5 communities show significantly higher average rental prices, indicating premium residential communities.
